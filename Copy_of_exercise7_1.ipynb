```python
def get_webpage_soup(url):
  headers = {
        "User-Agent": "Mozilla/5.0"
    }
  response = requests.get(url, headers=headers)
  if response.status_code != 200:
    print(f"Error: {response.status_code}")
    return None
  else:
    print(f"{url} was scraped successfully")
    return BeautifulSoup(response.text, 'html.parser')
```

```python
def scrape_newsbomb_teaser_pages(archive_name, start_page_nbr=1, end_page_nbr=1):
    all_articles_dataL = []
    base_url = f"https://www.newsbomb.gr/{archive_name}"

    print(f"\rScraping started for {archive_name}", end="")

    for page_nbr in range(start_page_nbr, end_page_nbr + 1):
        url = f"{base_url}?page={page_nbr}" if page_nbr > 1 else base_url
        print(f"\rScraping page {page_nbr}: {url.ljust(80)}", end="")

        soup = get_webpage_soup(url)
        if not soup:
            continue


        articles_container = soup.find("div", class_="page-section news-items-section") or \
                           soup.find("div", class_="articles-list") or \
                           soup.find("div", class_="news-items")

        if not articles_container:
            print(f"\nNo articles container found on page {page_nbr}")
            continue

        articlesL = articles_container.find_all("article", class_=lambda x: x and 'card' in x) or \
                   articles_container.find_all("article")

        if not articlesL:
            print(f"\nNo articles found on page {page_nbr}")
            continue

        page_articles_dataL = get_teaser_elements_from_articlesL(articlesL)
        all_articles_dataL.extend(page_articles_dataL)

        delay = random.uniform(1.5, 4)

        time.sleep(delay)

    all_articles_df = pd.DataFrame(all_articles_dataL)
    print(f"\rScraping completed. Fetched {len(all_articles_df)} articles from {archive_name}.".ljust(100))

    return all_articles_df
```
